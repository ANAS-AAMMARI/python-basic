```python
import tkinter as tk
from tkinter.filedialog import askopenfile
import cv2

root = tk.Tk()

file = askopenfile(mode='r', filetypes=[('Video Files', ["*.mp4","*.avi"])])
if file:
# #   global filename
# #   filename = file.name
    cap = cv2.VideoCapture(file.name)

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        cv2.imshow("Walking frame", frame)
        if cv2.waitKey(50) == 27:
            break
    else:
        break
else:
    print("Error opening video  file")
cv2.destroyAllWindows()
cap.release()
root.mainloop()
```

```python
from tkinter import *
from PIL import Image, ImageTk
import cv2

# Create an instance of TKinter Window or frame
win= Tk()

# Set the size of the window
win.geometry("700x350")# Create a Label to capture the Video frames
label =Label(win)
label.grid(row=0, column=0)
cap= cv2.VideoCapture("vid.avi")

# Define function to show frame
def show_frames():
    # Get the latest frame and convert into Image
    cv2image= cv2.cvtColor(cap.read()[1],cv2.COLOR_BGR2RGB)
    img = Image.fromarray(cv2image)
    # Convert image to PhotoImage
    imgtk = ImageTk.PhotoImage(image = img)
    label.imgtk = imgtk
    label.configure(image=imgtk)
# Repeat after an interval to capture continiously
label.after(20, show_frames)

show_frames()
win.mainloop()
```

```python
from tkinter import *
from tkinter.filedialog import askopenfile
from PIL import Image, ImageTk
import imutils
import cv2


def affiche():
    global cap
    if cap is not None:
        ret, frame = cap.read()
        if ret == True:
            frame = imutils.resize(frame, width=640)
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            im = Image.fromarray(frame)
            img = ImageTk.PhotoImage(image=im)
            lblVideo.configure(image=img)
            lblVideo.image = img
            lblVideo.after(10, affiche())
        else:
            lblVideo.image = ""
            cap.release()


def visualisation_video():
    global cap
    video_path = file = askopenfile(mode='r', filetypes=[('Video Files', ["*.mp4","*.avi"])])
    if video_path:
        lblInfoVideoPath.configure(text=video_path)
        cap = cv2.VideoCapture(video_path.name)
        affiche()
    else:
        lblInfoVideoPath.configure(text="aucun video selectione")
    
cap= None   
root = Tk()

btn_vis = Button(root, text="hhhkkkk", command=visualisation_video)
btn_vis.grid(column=0,row=0,padx=5, pady=5,  columnspan=2)

lblInfo1 = Label(root, text="video:")
lblInfo1.grid(column=0, row=1)

lblInfoVideoPath = Label(root, text="aucun video selectione")
lblInfoVideoPath.grid(column=1, row=1)

lblVideo = Label(root)
lblVideo.grid(column=0, row=2, columnspan=2)
root.mainloop()
```
